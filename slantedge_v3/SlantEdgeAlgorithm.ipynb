## Slant Edge Algorithm



## Introduction

This notebook presents the Slant Edge algorithm to compute the MTF (Modulation transfer function) to characterize a camera lens. It is presented step by step so that an understanding can be gained by reading this notebook. The notebook provides a lot of extra code which is used soley for understanding. The actual code would be shorter.

![](./mtf.jpg)


This diagram shows the MTF for the corresponding slant edge image.

![](./slantedge.png)

The MTF graph was obtained by executing the matlab script [srfmat3](http://losburns.com/imaging/software/SFRedge/sfrmat3_post/index.html) written by Peter D. Burns. The python script in this notebook is very much inspired from this code. It looks to be a safe reference to base off of, as the imatest software is also based on it.

The algorithm can be broken down into the following steps:

  1. (not implemented yet) gamma correction
     * This is an addition of the imatest software. The srfmat3 original does not implement this either.
  2. it a line on the edge
  3. Compute the average ESF (Edge Spread Function).
  4. Derivate to obtain the LSF (Line Spread Function).
  5. Do a discrete fourier transform. The magitude is the MTF

### Purpose

The purpose of this algorithm is to characterize the image taken by a camera. The image is never perfect on a picture. If we take a picture of a very far point light, like a star, it should in theory appear as a point, or a single pixel in the image. But it will often be more like a halo. This is the point spread function (PSF). An image is taking actually a lot of points, so by just adding the PSF for all the points, we can obtain the resulting image in the camera.

![](https://upload.wikimedia.org/wikipedia/commons/c/c2/Convolution_Illustrated_eng.png)

It is very useful here to go in the frequency domain because the convolution of the PSF becomes a simple multiplication in the frequency domain. What the MTF is, is the LSF which is a 1 dimensional slice of PSF, transformed in the frequency domain. This allows to see many interesting properties about the lens.

## ROI Rotation

The slant edge picture is no in the correction orientation on purpose. There is actually no "correct" orientation, but it is good to normalize its orientation for the further steps. The edge need to be vertical and go down left.

First the image is rotated so that the edge goes up-down instead of left-right. This is done by computing two values:

* **test_v** : the color value of the 3rd row from the beginning is averaged. The color from the 3rd row from the end is averaged. In the above image, it is expected to be black and white. The absolute difference of those two values is saved.
* **test_h** : the color value of the 3rd column from the beginning is averaged. Same with 3rd column from the end. Both colors are expected to be some form of gray. The absolute difference of those two values is saved.

So test_v is bigger than test_h, which means it is not correctly oriented.

All the calculations are done in one color channel. Here we use the green channel.

```python
nn = 3 # row/column index to take average from
mm = 1 # color channel

test_v = np.abs(np.mean(src[nn,:,mm]) - np.mean(src[-nn,:,mm]))
test_h = np.abs(np.mean(src[:,nn,mm]) - np.mean(src[:,-nn,mm]))

if test_v > test_h:
    src = cv.rotate(src, cv.ROTATE_90_CLOCKWISE)

plt.imshow(src[:,:,::-1])
plt.show()
```

Now, the white side should be on the left. We use the same techniques to determine if the orientation is correct.

```python
test_left = np.mean(src[:,nn,mm]) # Compute average of 3rd column from the beginning
test_right = np.mean(src[:,-nn,mm]) # Compute average of 3rd column from the end

if test_right > test_left: # If right side is white, flip
    src = cv.flip(src, 1) # flip around the y-axis

plt.imshow(src[:,:,::-1])
```

Finally the edge need to go down left. To detect this, we will compute the centroid of the edge on the top and bottom.

We will find the centroid (in 1D), also called center of mass in physics.

Take a row i and let $v_i(j)$ be the pixel value at column j. The centroid $c_i$ for the row i is defined as:

$c_i = \frac{\sum_j v_i(j) \cdot j}{\sum_j v_i(j)}$

```python
def centroid(row):
    j = np.arange(len(row)) # list of indices [0, 1, ..., len(row)-1]
    sumx = np.sum(row) # denominator of formula
    if sumx < 1e-4: # avoid division by zero
        c = 0
    else:
        c = np.sum(j * row)/sumx # centroid computation
    return c
```

```python
c1 = centroid(src[nn,:,mm])
c2 = centroid(src[-nn,:,mm])

print(f"centroid top: {c1}")
print(f"centroid bot: {c2}")


```

```python
if c2 > c1:
    src = cv.flip(src, 0) # Flip around x-axis

plt.imshow(src[:,:,::-1])
```

## Find edge

It is interesting to see that we could in theory already obtain the ESF (edge spread function) at this point. We just need to select a row of the image. 

```python
# Here the row 50 and the channel 0 (Blue) is arbitrary selected
plt.plot(src[50, :, 0])
```

The general idea of the slanted edge is to obtain better precision than that, we will average the ESF of all the rows of the image. But because it is slanted, we cannot just superimpose all the ESF on top of each other. Instead, we need to shift slightly for every next row. So the first goal will be to find the angle of the edge or its slope.

But first, we also have to make the edge more apparent. A common way to do this, is to calculate the first derivative of the image. Here, to show that, we can take the above graph of the ESF, and calculate the 1D gradient.

To obtain the gradient, the cv.filter2D which allows to convolve with any kernel.

```python
# Here the row 50 and the channel 0 (Blue) is arbitrary selected
lsf = cv.filter2D(np.array([src[50, :, 0]]), cv.CV_64F, np.array([[0.5, -0.5]]))
plt.plot(lsf[0])
```

A large spike is visible, this is where the edge is because the derivative is large. The same procedure is done row by row but for the entire image. As an example, the input will be the channel 1 (green).

```python
def deriv1(img):
    result = cv.filter2D(img, cv.CV_64F, np.array([[0.5, -0.5]]))
    return result

result = deriv1(src[:,:,1])
plt.imshow(result, cmap="gray")
```

We get a nice image of the edge. The goal now will be to fit a line onto it. This is not entirely straighforward because there is no such function in OpenCV. There is a [fitLine](https://docs.opencv.org/3.4/d3/dc0/group__imgproc__shape.html#gaf849da1fdafa67ee84b1e9a23b93f91f) but to my knowledge this cannot be used to fit onto an image but only a list of points. 

Instead, for **each row**, we find the centroid $c_i$.

Define a list of points $p_i = (i, c_i)$. This is the 2d coordinates points of the centroids for each row. 

Do a polynomional fitting of 1st degree onto the points $p_i$ to find the best line fit.

Additionally, we apply a hamming window on the line to obtain a smoother result. This is mainly used later when computing the fourier transform but it's also used here for consistency.

```python
def find_centroid(result):
    centroids = []

    for i in range(result.shape[0]): # Iterate over each row
        win1 = np.hamming(len(result[i,:]))
        c = centroid(result[i,:]*win1)
        c -= 0.5 # Minor correction for offset because of derivation
        centroids.append(c)
    return centroids

centroids = find_centroid(result) 

plt.plot(centroids, 'g.')
plt.grid('on')
plt.show()
```

## Fit line

Do a polynomial 1st degree fitting onto the points.

```python
def fit_line(centroids):
    i = np.arange(len(centroids))
    z = np.polyfit(i, centroids, 1)
    
    return z

z = fit_line(centroids)
print(z)
```

The first return value is the slope and the second the intercept. Plotting the line the line onto the points shows, it indeeds approximate it nicely.

```python
p = np.poly1d(z)

plt.plot(centroids, 'g.')
plt.plot(p(np.arange(len(centroids))), 'r--')
plt.grid('on')
plt.show()
```

The angle $\alpha$ of the slanted edge can be computed easily now with the slope.

$\alpha = atan(-slope)$

```python
import math

def compute_angle(z):
    return math.atan(-z[0]) * 180/np.pi

print(compute_angle(z)) # convert radians to degrees
```

The fitted line can also be superimposed onto the source image.

```python
annotated = cv.cvtColor(src[:,:,1], cv.COLOR_GRAY2BGR) # Take green channel and conver it to 3 channel to annotate in color
x0, y0 = p(0), 0 # Line point at the top
x1, y1 = p(src.shape[0]), src.shape[0] # Line point at the bottom

cv.line(annotated, (round(x0), round(y0)), (round(x1), round(y1)), (0, 0, 255), 1)
plt.figure(figsize=(10, 10))
plt.imshow(annotated[:,:,::-1])
plt.show()
```

## Compute ESF

Now, all the ESF are averaged into one. For this an accumulator is created with two lines. The first will contain the accumulated pixels values and the second the count. To get the average, the first line has be to be divided by the second line.

But this is too simple! srfmat3 had to go more complicated. It uses a supersampling method to slightly improve the resulting ESF more. With the pixels shifted for each row, the x position is a float number. Instead of applying a round/floor/ceil, the float position is kept somewhat by having an accumulator with more cells. 

Here we will use a supersampling of 4x, so a single unit is divided into 4. The accumulator has all to take into account that the pixels in the bottom rows could go outside the frame because of the shifting. The accumulator has to be extended to avoid any out of range error.

Taken all this into account, the accumulator array is created.

```python
fac = 4 # supersampling multiplier
green = src[:,:,1]
extra = math.ceil(abs(green.shape[0]*z[0])) # Extra to account for pixels that could go outside the frame
bins = np.zeros((2, (green.shape[1] + extra)*fac))

print(green.shape)
print(bins.shape)
```

The pixels of the input image are all processed and put into the accumulator array.

```python
for x in range(green.shape[1]):
    for y in range(green.shape[0]):
        x_off = math.ceil((x - z[0]*y)*fac) # computed shifted x position
        bins[0, x_off] += 1 # count 
        bins[1, x_off] += green[y, x] # pixel value
```

### Complete the zero accumulator cells

Some cell of the accumulator could be empty. This is problematic so the accumulator value on the left is taken and copied.

```python
for i in range(bins.shape[1]):
    if bins[0, i] == 0: # count is 0
        bins[0, i] = bins[0, i-1]
        bins[1, i] = bins[1, i-1]
```

Compute the average ESF.

```python
esf = bins[1, :] / bins[0, :]
plt.plot(esf)
plt.show()
```

The same complete function is written below

```python
def compute_esf(img, z, fac=4):
    extra = math.ceil(abs(img.shape[0]*z[0])) # Extra to account for pixels that could go outside the frame
    bins = np.zeros((2, (img.shape[1] + extra)*fac))

    for x in range(img.shape[1]):
        for y in range(img.shape[0]):
            x_off = math.ceil((x - z[0]*y)*fac) # computed shifted x position
            bins[0, x_off] += 1 # count 
            bins[1, x_off] += img[y, x] # pixel value
            
    for i in range(bins.shape[1]):
        if bins[0, i] == 0: # count is 0
            bins[0, i] = bins[0, i-1]
            bins[1, i] = bins[1, i-1]   
            
    esf = bins[1, :] / bins[0, :]
    return esf
```

## Compute LSF

From the ESF, the LSF is just its derivative. We use the same technique as before with the convolution but with a kernel of 1x3 with $\begin{bmatrix}-0.5  & 0 & 0.5 \end{bmatrix}$. We also have to be careful of the border which gets incorrect values. The value besides them is copied.

```python
 def get_lsf(esf):
    lsf = np.convolve(esf, np.array([-0.5, 0, 0.5]), 'same')
    lsf[0] = lsf[1]
    lsf[-1] = lsf[-2]
    
    return lsf

lsf = get_lsf(esf)

plt.plot(lsf)
plt.show()
```

Comparing with the LSF of a single row, it looks more refined and precise, so it was worth the effort after all.

Now, before computing the discrete fourier transform. The LSF is centered. The centroid formula is again used to find the center, and the plot is shifted.

```python
c = centroid(lsf) 
print(f"Center is at {len(lsf)/2}, centroid at {c}")
```

```python
d =  round(c) - len(lsf)//2
centered = np.zeros_like(lsf)
m = len(lsf)

if d > 0: # Shift left
    for i in range(m-d):
        centered[i] = lsf[i + d]
elif d < 1: # Shift right
    for i in range(-d, m):
        centered[i] = lsf[i + d]
else: # Already centered
      centered = lsf

plt.plot(lsf)
plt.plot(centered)
plt.legend(['original', 'centered'])
```

```python
def recenter(lsf):
    c = centroid(lsf)
    
    d =  round(c) - len(lsf)//2
    centered = np.zeros_like(lsf)
    m = len(lsf)

    if d > 0: # Shift left
        for i in range(m-d):
            centered[i] = lsf[i + d]
    elif d < 1: # Shift right
        for i in range(-d, m):
            centered[i] = lsf[i + d]
    else: # Already centered
        centered = lsf
    return centered
```

## Compute MTF

Finally, the discrete fourier transform of the MTF is computed. This is done by passing the LSF to the numpy's fft function.
The output is the complex frequency, the absolute value is taken which is return to reals and can be plotted. This is actually the MTF.

The frequency for each component is interesting. The smallest frequency is 1 pixel/cycle while the biggest frequency is the whole image/cycle. The frequency have to be adjusted slightly because the LSF is taken from a sloped line. The fundamental unit is not 1 anymore.

Also because of the supersampling, the frequency goes actually higher than 1 pixel/cycle. But here we limit the frequency to 0.75/cycle again. The higher frequency look inconsistent.

The MTF is also normalized with the base frequency to have a nice plot.

A hamming window is applied to the LSF to obtain a smoother result in the fourier domain.

```python
def get_mtf(centered, z, fac=4):
    # Apply hamming window
    win = np.hamming(len(centered))
    out = np.fft.fft(centered*win) # Compute FFT

    freqlim = 1 # limit frequency
    delfrac = math.cos(math.atan(z[0])) # fundamental unit of slope line = 0.995
    n = np.arange(len(centered))

    freq = n*fac/(len(centered)*delfrac) # fac is supersampling = 4

    nnout = np.sum(freq < 0.75) # number of entries with freq < 0.75

    mtf = np.abs(out[:nnout]) # Compute magnitude
    mtf = mtf/mtf[0] # Normalize
    
    return mtf, freq[:nnout]

mtf, freq = get_mtf(centered, z)

plt.plot(freq, mtf)
plt.grid('on')
plt.title('MTF')
plt.ylim([0, max(mtf)*1.1])
plt.legend(['green channel'])
plt.xlabel('frequency, cycles/pixel')
plt.ylabel('SRF/MTF')
plt.show()
```

## MTF50 and MTF50P

MTF50 is defined to be the frequency where the MTF is 50% of its low frequency value. The MTF50P is where the MTF is 50% of its peak value. Both values are important to define sharpness.

```python
def mtf50(mtf, freq):
    for i in range(len(mtf)):
        if mtf[i] <= 0.5:
            return freq[i]
    return None

print(mtf50(mtf, freq))
```

```python
def mtf50p(mtf, freq):
    max_mtf = np.max(mtf)
    for i in range(len(mtf)):
        if mtf[i] <= 0.5*max_mtf:
            return freq[i]
    return None

print(mtf50p(mtf, freq))
```

## MTF for all channels

In addition to the RGB channels, srfmat3 computes an additionnal channel which is just a weighted average of the RGB channel called luminance. This can be thought of the perceived luminance of all the channels combined. Green is better perceived by the retina so it receives more weight. The formula is as follows:

$Lum = 0.213 \cdot R + 0.715 \cdot G  + 0.072 \cdot B$

```python
# opencv stores images in BGR
lum = 0.213 * src[:,:,2] + 0.715 * src[:,:,1] + 0.072*src[:,:,0]

# make a new image with 4 channels for convenience
cc = np.array([src[:,:,2], src[:,:,1], src[:,:,0], lum])
```

Do the calculations for all the channels, display the MTF for all of them.

```python
angles = []

plt.figure(figsize=(15, 15))

styles = [
    'r:',
    'g-',
    'b--',
    'k^'
]

mtf50s = []
mtf50ps = []

for i in range(4):
    img = cc[i, :,:]
    # Line fitting
    result = deriv1(img)
    centroids = find_centroid(result)
    z = fit_line(centroids)
    
    angle = compute_angle(z)
    angles.append(angle)
    
    # LSF
    esf = compute_esf(img, z)
    lsf = get_lsf(esf)
    
    # MTF
    centered = recenter(lsf)
    mtf, freq = get_mtf(centered, z)
    
    plt.plot(freq, mtf, styles[i])
    
    # Compute statistics
    mtf50s.append(mtf50(mtf, freq))
    mtf50ps.append(mtf50p(mtf, freq))
    
plt.grid('on')
plt.title('MTF')
plt.ylim([-0.2, 1.2])
plt.legend(['red', 'green', 'blue', 'luminance'])
plt.xlabel('frequency, cycles/pixel')
plt.ylabel('SRF/MTF')
plt.text(0.4, 1, f"MTF50: {mtf50s[-1]:.3f}\nMTF50p: {mtf50ps[-1]:.3f}", bbox=dict(facecolor='red', alpha=0.5), fontsize=16)
plt.xticks(np.arange(0, 0.75, 0.05))
plt.yticks(np.arange(-0.2, 1.2, 0.05))
plt.savefig('mtf.png') 
plt.show()
```
